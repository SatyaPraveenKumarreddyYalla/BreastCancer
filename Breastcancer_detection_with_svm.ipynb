#LIberies/Modues

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn import svm # Corrected import: SVM models are in sklearn.svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

#dATA PROCESSING

```python
df = pd.read_csv("/content/sample_data/breast-cancer.csv")
df.head()
```

```python
df.info()
df.dtypes
df.isnull().sum()
```

```python
df.describe()
```

```python
df1 = df.copy()
le = LabelEncoder()
df1['diagnosis'] = le.fit_transform(df1['diagnosis'])
df1.head()
```

```python
x = df1.drop(columns='diagnosis', axis=1)
y = df1['diagnosis']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.8,random_state=48)
paramgrid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf'],
    'random_state': [48],
    'probability': [True],
    'class_weight': ['balanced'],
    'decision_function_shape': ['ovr'],
    'max_iter': [-1],
    'decision_function_shape': ['ovr'],
    'break_ties': [False],
    'verbose': [False],
    'cache_size': [200],
    'shrinking': [True],
    'tol': [0.001],
    'decision_function_shape': ['ovr'],
    'break_ties': [False],
    'verbose': [False],
    'cache_size': [200],
    'shrinking': [True],
    'tol': [0.001],
    'decision_function_shape': ['ovr'],
}
grid = GridSearchCV(svm.SVC(), paramgrid, refit=True, verbose=3)
grid.fit(xtrain, ytrain)
grid_predictions = grid.predict(xtest)
print(" Best Parameters result: ",grid.best_params_)
print("Classification report: ",classification_report(ytest,grid_predictions))
print("Confusion Matrix: ",confusion_matrix(ytest,grid_predictions))
fig = px.scatter(df, x="radius_mean", y="texture_mean", color="diagnosis", hover_data=['diagnosis'])
fig.update_layout(
    title="Breast Cancer Diagnosis",
    xaxis_title="Radius Mean",
    yaxis_title="Texture Mean",
    legend_title="Diagnosis",
    hovermode="closest",
    plot_bgcolor="white",
    paper_bgcolor="white",
    font=dict(family="Arial", size=12, color="black"),
    margin=dict(l=50, r=50, t=50, b=50),
    xaxis=dict(showgrid=True, gridwidth=0.5, gridcolor="lightgray"),
    yaxis=dict(showgrid=True, gridwidth=0.5, gridcolor="lightgray"),
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01)
)
fig.show()
```
